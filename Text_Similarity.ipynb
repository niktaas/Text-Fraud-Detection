```python
def deEmojify(text):
    regrex_pattern = re.compile(pattern = "["
        u"\U0001F600-\U0001F64F"  # emoticons
        u"\U0001F300-\U0001F5FF"  # symbols & pictographs
        u"\U0001F680-\U0001F6FF"  # transport & map symbols
        u"\U0001F1E0-\U0001F1FF"  # flags (iOS)
                           "]+", flags = re.UNICODE)
    return regrex_pattern.sub(r'',text)

def remove_urls(text):
    url_pattern = re.compile(r'https?://\S+|www\.\S+')
    return url_pattern.sub(r'', text)

def remove_html(text):
    html_pattern = re.compile('<.*?>')
    return html_pattern.sub(r'', text)

def remove_html(text):
    html_pattern = re.compile('<.*?>')
    return html_pattern.sub(r'', text)

def remove_hashtag(text):
    return re.sub(r'#([^\s]+)', r'\1', text)

def remove_extra_repeated_alpha(text):
    return re.sub(r'([^\W\d_])\1{2,}', r'\1', text)

def remove_punctuations(text):
    punctuations = re.compile(r'[^\w\s]')
    return punctuations.sub(r'', text)

def remove_latin(text):
  output = re.sub(r'[A-Za-z]+','', text)
  output = re.sub(r'[0-9]+','', output)
  #output = re.sub(r"^[ آابپتثجچحخدذرزژسشصضطظعغفقکگلمنوهی]+$", "", text)
  output.rstrip()
  return output
```

```python
lemma = InformalLemmatizer()
normalizer = Normalizer()

def clean_up(text,norm =False, latin = False, emoji = True, url=True, html=True, punctuations=True, lemmatize=False, hashtag=True, stopwords=True, remove_extra_alpha=True):
    # remove emoji
    if emoji:
      text = deEmojify(text)
    # remove url
    if url:
        text = remove_urls(text)
    # remove html tags
    if html:
        text = remove_html(text)
    # remove punctuations
    if punctuations:
        text = remove_punctuations(text)
    # Alter words with repeated alphabets
    if remove_extra_repeated_alpha:
        text = remove_extra_repeated_alpha(text)
    # remove hashtag
    if hashtag:
      text = remove_hashtag(text)
    #remove latin and number
    if latin:
      text = remove_latin(text)
    # remove _
    text = text.replace("_", "")
    # text = text.replace("¾ص", "")
    # text = text.replace("çııı", "")
    # text = text.replace("çə", "")
    # text = text.replace("ö", "")
    # text = text.replace("ø", "")
    # text = text.replace("üü", "")
    # text = text.replace("üşğ", "")
    # text = text.replace("üəə", "")
    # text = text.replace("ğı", "")
    # text = text.replace("ıö", "")
    # text = text.replace("ıı", "")
    # text = text.replace("şı", "")
    # text = text.replace("şış", "")
    # text = text.replace("şəə", "")
    # text = text.replace("əüə", "")
    # text = text.replace("əı", "")
    # text = text.replace("əə", "")
    # text = text.replace("əəi", "")
    # text = text.replace("əəə", "")
    # text = text.replace("ə", "")
    # text = text.replace("ş", "")
    # text = text.replace("ğ", "")
    # text = text.replace("ü", "")
    # text = text.replace("ı", "")
    if norm:
      text = normalizer.normalize(text)
    # normalize text

    # tokenize text
    tokens = word_tokenize(text)
    # remove stop words
    if stopwords:
      tokens = [word for word in tokens if word not in sw]
    # lemmatize words
    if lemmatize:
      tokens = [lemma.lemmatize(word) for word in tokens]
    #text = ' '.join(tokens)


    return tokens
```

```python
def jaccard_similarity(x,y):
  """ returns the jaccard similarity between two lists """
  intersection_cardinality = len(set.intersection(*[set(x), set(y)]))
  if intersection_cardinality/len(set(x)) >= 0.8 and intersection_cardinality/len(set(y))<=0.8:
    print('x is subset')
    return intersection_cardinality/len(set(x))
  elif intersection_cardinality/len(set(y)) >= 0.8 and intersection_cardinality/len(set(x))<=0.8:
    print('y is subset')
    return intersection_cardinality/len(set(y))
  else:
    union_cardinality = len(set.union(*[set(x), set(y)]))
    return intersection_cardinality/float(union_cardinality)
```
